```python
#1. Replace NaN with 0 and interchange rows & columns
import numpy as np

a = np.array([[6, -8, 73, -110],
              [np.nan, -8, 0, 94]])

# Replace NaN with 0
a = np.nan_to_num(a, nan=0)

print("After replacing NaN:")
print(a)

# Interchange rows
a[[0,1]] = a[[1,0]]
print("After row interchange:")
print(a)

# Interchange columns 0 and 3
a[:,[0,3]] = a[:,[3,0]]
print("After column interchange:")
print(a)
```

```python
#2) Move axes of 3D array to new positions
import numpy as np

a = np.arange(24).reshape(2,3,4)

print("Original Shape:", a.shape)

b = np.moveaxis(a, 0, 2)

print("New Shape:", b.shape)
print(b)
```

```python
#3) Replace NaN values with average of columns
import numpy as np

a = np.array([[1,2,np.nan],
              [4,np.nan,6],
              [7,8,9]])

col_mean = np.nanmean(a, axis=0)

inds = np.where(np.isnan(a))
a[inds] = np.take(col_mean, inds[1])

print(a)
```

```python
#4) Replace negative values with zero
import numpy as np

a = np.array([4,-5,7,-2,-9,10])

a = np.where(a < 0, 0, a)

print(a)
```

```python
#5) Mean, Median and Average of two 2D Arrays

import numpy as np

arr1 = np.array([[1,2,3],
                 [4,5,6]])

arr2 = np.array([[7,8,9],
                 [10,11,12]])

combined = np.concatenate((arr1, arr2))

print("Combined Array:")
print(combined)

print("Mean =", np.mean(combined))
print("Median =", np.median(combined))

# Average
avg = (arr1 + arr2)/2
print("Average Array:")
print(avg)

from scipy import stats
import numpy as np

combined1 = np.array([1,2,2,3,4,5,5,5])

mode = stats.mode(combined1)

print("Mode =", mode.mode)
```

```python
#6) Solve equations using linalg() and inverse matrix

import numpy as np

A = np.array([[1,-2,3],
              [-1,3,-1],
              [2,-5,5]])

B = np.array([9,-6,17])

# Method 1
X = np.linalg.solve(A,B)

print("Using linalg.solve:")
print(X)

# Method 2 (Inverse Matrix)
A_inv = np.linalg.inv(A)

X2 = np.dot(A_inv,B)

print("Using Inverse Matrix:")
print(X2)
```

```python
#1. Python function to find the maximum of three numbers
def maximum(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(maximum(10, 25, 15))
```

```python
#2. Python function to return distinct elements from a list
def distinct_list(lst):
    new_list = []

    for i in lst:
        if i not in new_list:
            new_list.append(i)

    return new_list

print(distinct_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
#3. Python function to multiply all the numbers in a list
def multiply_list(lst):
    result = 1

    for i in lst:
        result = result * i

    return result

print(multiply_list([2, 3, 4, 5]))
```

```python
#4. Python function to calculate the factorial of a number
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

print(factorial(5))
```

```python
#5. Python function to reverse a string
def reverse_string(s):
    return s[::-1]

text = input("Enter a string: ")
print("Reversed string is:", reverse_string(text))
```

```python
#6. Python function to check whether a number is in a given range
def check_range(num, start, end):
    if num >= start and num <= end:
        print(num, "is within the range")
    else:
        print(num, "is outside the range")

check_range(5, 1, 10)
```

```python
#7. Python function to print even numbers from a given list
def even_numbers(lst):
    for i in lst:
        if i % 2 == 0:
            print(i)

numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_numbers(numbers)
```

```python
#8. Python function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
#9. Python function to count uppercase and lowercase letters in a string
def count_case_letters(text):
    upper = 0
    lower = 0

    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case_letters("Hello World")
```

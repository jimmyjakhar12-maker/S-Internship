```python
#1. Create NumPy Array and Convert 1D Array to 2D
import numpy as np
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

arr2d = arr.reshape(2, 3)

print("Original Array:")
print(arr)

print("\n2D Array:")
print(arr2d)
```

```python
#2. Print Array Attributes
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Data Type:", arr.dtype)
print("Item Size:", arr.itemsize)
```

```python
#3. Create a 3×3 NumPy Array of All 9
arr = np.full((3, 3), 9)
print("3×3 Array of All 9:")
print(arr)
```

```python
#4. Create a 1D Array of 10 Evenly Spaced Values Between 25 and 125
import numpy as np

arr = np.linspace(25, 125, 10)

print(arr)
```

```python
#5. Convert a Python List into a NumPy Array
import numpy as np

lst = [10, 20, 30, 40, 50]

arr = np.array(lst)

print(arr)
```

```python
#6. Reverse a 1D NumPy Array
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

rev = arr[::-1]

print(rev)
```

```python
#7. Create a 4×4×3 Array and Extract Value at Second Set, First Row and Last Column

import numpy as np

arr = np.arange(48).reshape(4, 4, 3)

print(arr)

value = arr[1, 0, 2]

print("\nRequired Value =", value)
```

```python
#8. Create a 4×4 Array and Extract Odd Rows and Even Columns
import numpy as np

arr = np.arange(1, 17).reshape(4, 4)

print("Original Array:")
print(arr)

result = arr[::2, 1::2]

print("\nOdd Rows and Even Columns:")
print(result)
```

```python
#9. Slice the First Two Rows and First Two Columns of Second Set from a 4×4×3 Array
import numpy as np

arr = np.arange(48).reshape(4, 4, 3)

result = arr[1, :2, :2]

print(result)
```

```python
#10. Replace All Odd Numbers with -1 Using For Loop
import numpy as np

arr = np.array([[23, 56, 78, 93],
                [71, 82, 13, 24]])

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] % 2 != 0:
            arr[i][j] = -1

print(arr)
```

```python
#11. Get Indices of Non-Zero Elements
import numpy as np

arr = np.array([1, 0, 2, 0, 3, 0, 4])

indices = np.nonzero(arr)

print(indices)
```

```python
#12. Add Two NumPy Arrays Element-wise
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([5, 15, 25])

result = arr1 + arr2

print(result)
```

```python
#13. Multiply Two NumPy Arrays Element-wise
import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([5, 15, 25])

result = arr1 * arr2

print(result)
```

```python
#14. Compute Dot Product
import numpy as np

arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

dot_product = np.dot(arr1, arr2)

print("Dot Product =", dot_product)
```

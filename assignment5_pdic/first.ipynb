```python
#1. Repeat a tuple three times using *
t = (1, 2, 3 ,4)
result = t * 3
print(result)
```

```python
#2. Join three tuples using +
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

result = t1 + t2 + t3
print(result)
```

```python
#3. Check whether an element exists in a tuple
t = (10, 20, 30, 40 , 50)

element = 20

if element in t:
    print("Element exists")
else:
    print("Element does not exist")
```

```python
#4. Find total, highest, and lowest value without sum(), max(), min()

t = (14, 2, 21, 8, 11)

total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
#5. Filter tuple values greater than 10
n = (1, 14, 7, 28, 7, 39, 18, 34)

result = ()

for num in n:
    if num > 10:
        result += (num,)

print(result)
```

```python
#6. Count elements in a set without len()
s = {"cat", "dog", "bird", "fish"}

count = 0

for item in s:
    count += 1

print("Number of elements:", count)
```

```python
#7. Combine two sets
s1 = {1, 2, 3}
s2 = {3, 4, 5 ,6}

result = s1 | s2

print(result)
```

```python
#8. Find common elements in two sets
s1 = {1, 2, 3, 4, 5}
s2 = {3, 4, 5, 6}

common = s1 & s2

print(common)
```

```python
#9. Find all elements that are in either set (Union)
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1.union(s2)

print(result)
```

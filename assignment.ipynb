```python
# Student Marks Calculator

print("=" * 45)
print("       STUDENT MARKS CALCULATOR")
print("=" * 45)

name  = input("\nEnter Student Name  : ")
cls   = input("Enter Class         : ")

print("\nEnter marks for 5 subjects (out of 100):")
subject_names = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
marks = []

for subject in subject_names:
    while True:
        try:
            mark = float(input(f"  {subject} : "))_
            if 0 <= mark <= 100:
                marks.append(mark)
                break
            else:
                print("  ⚠  Please enter a value between 0 and 100.")
        except ValueError:
            print("  ⚠  Invalid input. Please enter a numeric value.")

total_marks     = sum(marks)
max_marks       = 500         
percentage      = (total_marks / max_marks) * 100

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

result = "PASS" if percentage >= 35 else "FAIL"

print("\n" + "=" * 45)
print("              RESULT CARD")
print("=" * 45)
print(f"  Student Name  : {name}")
print(f"  Class         : {cls}")
print("-" * 45)
print(f"  {'Subject':<12} {'Marks':>10} / 100")
print("-" * 45)
for subject, mark in zip(subject_names, marks):
    print(f"  {subject:<12} {mark:>10.1f}")
print("-" * 45)
print(f"  {'Total Marks':<12} {total_marks:>10.1f} / {max_marks}")
print(f"  {'Percentage':<12} {percentage:>9.2f}%")
print(f"  {'Grade':<12} {'':>10}{grade}")
print(f"  {'Result':<12} {'':>10}{result}")
print("=" * 45)
```

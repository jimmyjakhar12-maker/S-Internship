```python
#1 Create a csv file
import csv

# Dummy data entered by user
data = [
    ["Jitu Jaat", "Jhunjhunu, Rajasthan", "98765545210", "jitu23@gmail.com"],
    ["Kalu ", "Kerala, India", "9123236766", "kalu32@gmail.com"],
    ["Maninder", "Kota, Rajasthan", "9388773433", "manu34@gmail.com"]
]

# CSV file name
filename = "address_book.csv"

# Writing data into CSV file
with open(filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Address", "Mobile", "Email"])

    writer.writerows(data)

print("CSV file created successfully!")
```

```python
#2.Create database

import sqlite3

# Connect to database (creates database if not exists)
conn = sqlite3.connect("college.db")

# Create cursor object
cursor = conn.cursor()

# Student Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    city TEXT
)
""")

# Course Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_id INTEGER PRIMARY KEY AUTOINCREMENT,
    course_name TEXT,
    fees INTEGER
)
""")

# Teacher Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS teachers (
    teacher_id INTEGER PRIMARY KEY AUTOINCREMENT,
    teacher_name TEXT,
    subject TEXT
)
""")

print("Tables created successfully!")

# Insert into students
cursor.execute("INSERT INTO students (name, age, city) VALUES ('Amit', 21, 'Jaipur')")
cursor.execute("INSERT INTO students (name, age, city) VALUES ('Priya', 22, 'Delhi')")
cursor.execute("INSERT INTO students (name, age, city) VALUES ('Rahul', 20, 'Mumbai')")

# Insert into courses
cursor.execute("INSERT INTO courses (course_name, fees) VALUES ('Python', 15000)")
cursor.execute("INSERT INTO courses (course_name, fees) VALUES ('Java', 18000)")

# Insert into teachers
cursor.execute("INSERT INTO teachers (teacher_name, subject) VALUES ('Mr. Sharma', 'Python')")
cursor.execute("INSERT INTO teachers (teacher_name, subject) VALUES ('Ms. Verma', 'Java')")

conn.commit()

print("Records inserted successfully!")


print("\n--- All Students ---")
cursor.execute("SELECT * FROM students")

for row in cursor.fetchall():
    print(row)

print("\n--- Students from Jaipur ---")
cursor.execute("SELECT * FROM students WHERE city='Jaipur'")

for row in cursor.fetchall():
    print(row)

print("\n--- All Courses ---")
cursor.execute("SELECT * FROM courses")

for row in cursor.fetchall():
    print(row)


cursor.execute("""
UPDATE students
SET city='Kota'
WHERE name='Maninder'
""")

conn.commit()

print("\nStudent data updated successfully!")

print("\n--- Updated Students Table ---")
cursor.execute("SELECT * FROM students")

for row in cursor.fetchall():
    print(row)


cursor.execute("""
DELETE FROM students
WHERE name='Kalu'
""")

conn.commit()

print("\nStudent record deleted successfully!")

print("\n--- Final Students Table ---")
cursor.execute("SELECT * FROM students")

for row in cursor.fetchall():
    print(row)


conn.close()

print("\nDatabase connection closed!")
```

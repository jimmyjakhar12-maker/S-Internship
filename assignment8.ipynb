```python
import pandas as pd

# QUES 1 Convert a series of date-strings to a timeseries?
data = pd.Series([
    "12-07-2026","23-11-2000","02-12-2025"
])

print(data)

res = pd.to_datetime(data, dayfirst=True)
print(res)
print(type(res))
```

```python
# QUES 2 Create two DataFrames, df1 and df2, with a common column (e.g., 'ID').
df1 = pd.DataFrame({
    "id":[1,2,3,4],
    "Name":["A1","A2","A3","A4"],
    "sub":["eng","maths","phy","chemistry"]
})

df2 = pd.DataFrame({
    "id":[2,1,3,4],
    "Name":["B1","B2","B3","B4"],
    "sub":["python","java","c++","c+"],
    "marks":[12,00,18,15]
})

res = df1.merge(df2, on= "id", how="inner")
print(res)
```

```python
# Perform a left join of df1 and df2 on the 'ID' column.
df1 = pd.DataFrame({
    "id":[1,2,3,4],
    "Name":["A1","A2","A3","A4"],
    "sub":["eng","maths","phy","chemistry"]
})

df2 = pd.DataFrame({
    "id":[2,1,3,4],
    "Name":["B1","B2","B3","B4"],
    "sub":["python","java","c++","c+"],
    "marks":[12,00,18,15]
})

res = df1.join(df2,lsuffix="left",rsuffix="right",on="id", how="left")
print(res)
```

```python
# Perform a right join using pd.merge() on a common column

df1 = pd.DataFrame({
    "id":[1,2,3],
    "Name":["A","B","C"]
})

df2 = pd.DataFrame({
    "id":[2,3,4],
    "Marks":[80,90,70]
})

# Right Join
res = pd.merge(df1, df2, on="id", how="right")

print(res)


# perform a join using df.join() based on the index. Compare the results. Merging with Multiple Keys.

df1 = pd.DataFrame({
    "Name":["A","B","C"]
}, index=[1,2,3])

df2 = pd.DataFrame({
    "Marks":[84,34,75]
}, index=[2,3,4])

# Join on index
res2 = df1.join(df2, how="right")

print(res2)
```

```python
# Create three DataFrames. Vertically concatenate two of them using pd.concat()

df1 = pd.DataFrame({
    "id": [1, 2],
    "Name": ["Amal", "Ram"]
})

df2 = pd.DataFrame({
    "id": [3, 4],
    "Name": ["Neha", "Shiv"]
})

df3 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "Marks": [85, 90, 78, 88]
})

res = pd.concat([df1,df2],keys = ["x","y"], axis = 0)

print(res,"\n")
#  merge the resulting DataFrame with the third DataFrame on a common key. T Understand join() vs. merge().


res2 = pd.merge(res,df3,on="id",how="inner")
print(res2)
```
